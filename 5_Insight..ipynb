# **An Exploratory Analysis of Payroll and Overtime Trends in Louisville Metro Government**

# **Step 5: Summary and Recommendations**

# **5.1 Overview**
This section summarizes the key insights derived from the exploratory data analysis (EDA) of the Louisville Metro Government payroll dataset. The analysis focuses on understanding salary distribution, overtime patterns, departmental variations, and temporal trends using descriptive statistics and visual exploration techniques.



# **5.2 Primary Finding**
Payroll variation is mainly influenced by overtime utilization and job-role demand, while base salary alone does not fully explain payroll differences. The analysis shows that a small percentage of employees and departments contribute disproportionately to overtime costs, making overtime management the most effective cost-control strategy.



# **5.3 Key Metrics**
Total payroll records analyzed: ~40,000 employees

Salary distribution pattern: Right-skewed

Overtime concentration: Limited employees contribute majority of overtime pay

Salary growth trend: Gradual annual increase across years

Correlation insight: Weak relationship between base salary and overtime pay



# **5.4 Most Surprising Findings**
Higher salary does not guarantee higher overtime earnings

Median salary remains stable despite fluctuations in average salary

Department size does not always reflect overtime efficiency

Overtime spikes are caused by specific job roles rather than entire departments

# **5.5 Key Insights from the Employee Salary Analysis**

# **5.5.1 Salary Distribution Patterns**
The Annual_Rate values show a wide variation in employee compensation. Most employees fall within a moderate salary range, while a small number of employees earn significantly higher salaries. These high salaries are typically associated with senior management or specialized roles.

üìå Insight: Salary distribution is right-skewed, meaning a few employees earn much more than the majority.

# **5.5.2 Overtime Pay Trends**
Overtime compensation varies significantly across employees and departments. Certain departments consistently show higher overtime pay due to labor-intensive work, staffing shortages, or operational demands.

üìå Insight: Overtime pay is concentrated in specific departments rather than evenly distributed.

# **5.5.3 Department-Wise Salary Insights**
Departments differ considerably in average salary levels. Some departments have higher average salaries due to technical or senior job roles, while others rely more on overtime payments.

üìå Insight: Departmental salary differences highlight variations in job responsibilities and pay structures.



# **5.5.4 Relationship Between Annual Salary and Overtime Pay**
A weak-to-moderate correlation exists between Annual_Rate and Overtime_Pay. Employees with lower base salaries often receive higher overtime pay, possibly to compensate for workload. High-salary employees usually receive less overtime.

üìå Insight: Overtime is more common among mid-to-lower salary employees.

# **5.5.5 Anomalies and Outliers**
Several anomalies were identified, including extremely high salary values and unusually large overtime payments for a few employees. These cases may indicate exceptional workloads or policy variations.

üìå Insight: Outliers require further investigation for payroll auditing and policy review.

# **‚úÖ 5.6 Overall Findings**
The analysis reveals that employee compensation is influenced by department, job role, and overtime structure. While most employees earn within a standard salary range, overtime pay and salary outliers create significant payroll variation across the organization.

**Key Conclusions:**
Salary distribution is uneven, with a small number of high earners.

Overtime is department-specific and strongly impacts total compensation.

Compensation patterns reflect operational and workforce differences across departments.
